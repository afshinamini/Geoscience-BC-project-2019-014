```python
import lasio
import pandas as pd
from pandas import DataFrame
import os

# Libraries for data visualization:
import matplotlib.pyplot as plt
import seaborn as sns
```

## Export Gamma Ray reading within Montney Formation

```python
# Read the Files

las_file_list = []
path = 'Well_Logs/'

wellinfo_df = pd.read_csv('Well_ids_With_Montey_Tops.csv')
wellinfo_df.head()
```

```python
# View contents of the path, we will see LAS files
files = os.listdir(path)
files
```

```python
for file in files:
    if file.lower().endswith('.las'):
        las_file_list.append(path + file)
        
las_file_list
```

```python
df_list = []

for lasfile in las_file_list:
    las = lasio.read(lasfile)
    if 'UWI' in las.well:
        if 'GR' in las.curves:
            if 'DEPT' in las.curves:
                WellID = las.well.UWI.value
                if WellID in wellinfo_df['UWI'].values:
                    lasdf = las.df()
                    lasdf = lasdf.reset_index()
                    lasdf['GR'] = lasdf['GR'].fillna(0)
                    #lowerbound = wellinfo_df.loc[wellinfo_df['UWI'] == las.well.UWI.value,'Montney_Top']
                    #upperbound = wellinfo_df.loc[wellinfo_df['UWI'] == las.well.UWI.value,'Montney_Bottom']
            
                    df1 = wellinfo_df.loc[wellinfo_df['UWI'] == las.well.UWI.value]
                    lowerbound = df1['Montney_Top'].iloc[0]
                    upperbound = df1['Montney_Bottom'].iloc[0]
                    df = lasdf[lasdf['DEPT'].between(lowerbound, upperbound)]
                    Gamma = df['GR'].mean()
                    WellID = las.well.UWI.value
    df_list.append([WellID, Gamma])
#data.append([WellID, Gamma])
Gammadf = DataFrame (df_list,columns=['UWI','GR'])
#df = pd.DataFrame(data, columns=['WellID', 'Gamma'])            
```
